```python
with open('input.txt') as f:
    data = f.read().split()
len(data)
```

```python
def read_node(data):
    num_childs = int(data.pop(0))
    num_metadata = int(data.pop(0))
    assert(num_metadata > 0)

    nodes = []
    for x in range(num_childs):
        node = read_node(data)
        nodes.append(node)
        
    meta_data = []
    for x in range(num_metadata):
        meta = int(data.pop(0))
        meta_data.append(meta)
        
    return (nodes, meta_data)
    

data_copy = list(data)
head = read_node(data_copy)
head
```

```python
# Part 1
```

```python
def walk_nodes(node):
    yield node
    for child in node[0]:
        yield from walk_nodes(child)

meta_sum = 0
for node in walk_nodes(head):
    for x in node[1]:
        meta_sum += x
        
meta_sum
```

```python
# Part 2
```

```python
def node_value(node):
    children, metadata = node
    if len(children) == 0:
        return sum(metadata)

    value = 0
    for i in metadata:
        if i == 0:
            continue
        if i > len(children):
            continue
        child = children[i-1]
        node_val = node_value(child)
        value += node_val
        
    return value

    
    
node_value(head)
```
